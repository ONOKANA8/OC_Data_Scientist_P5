```python
# dÃ©finition d'une fonction d'obtention de rfm pour chaque pÃ©riode


def rfm(data):
    """
    Retourne tableau rfm.
    ParamÃ¨tres:
    data -- tableau contenant au moins les variables suivantes
    'order_id','customer_unique_id',
    'payment_value_1'->  montant de paiement total
    par 'customer_unique_id' (monetary),
    'order_purchase_timestamp' transformÃ© -> type datetime64.
    """

    separ = data.order_purchase_timestamp.astype(str)
    separ = separ.str.split(' ', n=1, expand=True).iloc[:, :-1]

    data["date_order_purchase_timestamp"] = separ

    # typage en datetime64
    dt = data.date_order_purchase_timestamp.astype("datetime64")
    data.date_order_purchase_timestamp = dt

    # temps max du dataset
    T_max = data.date_order_purchase_timestamp.max()
    group = data.groupby('customer_unique_id')
    dif_date = T_max-group["date_order_purchase_timestamp"].max()

    pd_dif = pd.DataFrame({"dif_date": dif_date})

    list_recency = pd_dif.dif_date.astype(str)
    list_recency = list_recency.str.split('days', n=1, expand=True)
    list_recency = list_recency.iloc[:, :-1][0].astype("int64").values

    list_frequency = group["order_id"].count().values
    list_monetary = group["payment_value_1"].sum().values
    list_customer_unique_id = group["payment_value_1"].sum().index

    rfm = pd.DataFrame({"customer_unique_id": list_customer_unique_id,
                        "Recency_days": list_recency,
                        "Frequency": list_frequency,
                        "Monetary": list_monetary})

    return rfm
```
